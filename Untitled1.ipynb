```python
# To help us calcualte cramers_v from above reference
# A 1 means highly associated, a 0 means not
def cramer_v(x,y):
    import scipy.stats as ss
    cofusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contigency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
    
```

```python
cramers_v(df['Outcome'], df['Outcome'])
```
